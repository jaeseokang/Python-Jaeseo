# [주제 및 데이터]

### 주제 : Burn it to Earn it : EDA + ML Models

칼로리 소모 예측을 위한 탐색적 데이터 분석(EDA)과 머신러닝 모델링 과정을 다룸.

### 데이터
- train.csv : 테스트 세트, 숫자형 변수 8개, 문자형 변수 1개, 변수명은 'id, Sex, Age, Height, Weight, Duration, Heart_Rate, Body_Temp, Calories), Null값 없음
- test.csv : 훈련 세트, 각각의 행에서 'Calories'를 예측함이 목표
- sample_submission.csv : correct format으로 존재
- 주요 input data에는 다음과 같은 feature가 포함됨(Gender, Age, Height, Weight, Duration, Heart Rate, Body temperature, Calories)

### 목적 
- 신체 활동 데이터 기반으로 칼로리 소모량을 정확하게 예측하는 모델 개발

# [코드 리뷰]

### (1) EDA
- data 불러오기 : kaggle 링크를 통해 train.csv와 test.cvs, sample_sub.csv를 불러옴.
- 범주형 변수(float 6개, int 2개), 문자형 변수(Sex) 확인
- 별도의 결측치 없음 확인
- Calories Distribution histplot 확인(right-skewed)
- Sex Distribution value_counts() 확인 : 남녀 비율 거의 동일
- Numerical Cols Distribution : 범주형 변수의 개별 subplot/histplot/scatter plot/line plot/violinplot/kdeplot 확인
- Sex 별 각 변수 mean 값 계산
- Numerical val의 corr를 heatmap으로 확인
- LabelEncoder 진행

### (2) 기능 엔지니어링 및 데이터 정리
- LabelEncoder한 변수값을 다시 train data에 대입
- train/test split : 종속변수 'Calories' 기준으로 데이터 분할

### (3) 모델 훈련 및 생성
- objective 함수 정의 : 파라미터(iterations, learning_rate, depth, 12_leaf_reg, bagging_temperature, border_count, loss_function, eval_metric, random_Seed, early_stopping_rounds, verbose, task_type)
- **CatBoostRegressor** 모델 적용 -> 범주형 변수에 강함, 순서 문제 해결, 적은 전처리 요구
- 모델 예측 정확도는 평균 절대 오차(MAE)와 같은 지표 통해 평가됨

### (4) 후처리
- best parameters를 이용해 모델 예측 후 rmsle 정의 후 최종 확인
- id별 Calories를 'submission.csv'로 제출

# [차별 점 및 배울 점]

### [차별 점]
- 범주형 변수에 최적화된 'CatBoostRegressor' 모델 사용
- 단순히 train, test dataset만 불러오는 것이 아니라 최종적으로 target value를 df로 나타내 'submission.csv'로 내보냄
- 최적의 parameters를 찾아내기 위해 GridSearchCV를 이용함

### [배울 점]
- 변수 중 8/9개가 범주형 변수로 이에 최적화된 'CatBoostRegressor'를 이용한게 인상깊었음
- 범주형 변수 plot를 여러가지 형태로 표현한 점이 인상깊었음.
